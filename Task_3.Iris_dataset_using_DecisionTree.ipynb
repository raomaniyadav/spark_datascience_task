```python
# Importing all libraries required in this notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load isir data set
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
iris["target_names"]
```

```python
feature= pd.DataFrame(iris["data"],columns=iris["feature_names"])
target=pd.Series(iris["target"])
```

```python
feature.head()
```

```python
target[:5]
```

#### find relationship between data

```python
feature.columns=['sepal_lenght','sepal_width','petal_length','petal_width']
```

```python
sns.pairplot(feature)
plt.show()
```

#### prepareing data 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,target)
```

#### Training the Algorithm
##### We have split our data into training and testing sets, and now is finally the time to train our algorithm.

```python
from sklearn.tree import DecisionTreeClassifier,tree
```

```python
model=DecisionTreeClassifier(criterion="gini",splitter="best",max_depth=2)
```

```python
model.fit(x_train,y_train)
```

#### evolution our model

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
y_act=y_test
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_act,y_pred)
```

```python
print(classification_report(y_act,y_pred))
```

```python
confusion_matrix(y_act,y_pred)
```
